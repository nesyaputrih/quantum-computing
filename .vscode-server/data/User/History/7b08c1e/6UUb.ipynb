```python
from qiskit import QuantumCircuit
from qiskit.circuit.library import ZGate
```

```python
n = 8 #qubit
gc = QuantumCircuit(n)
```

```python
#put h in any qubit
gc.h(range(n))
gc.draw("mpl")
```

```python
#oracle yg cuma memeriksa beberapa bit
#target 11001001
mcz = ZGate().control(num_ctrl_qubits = 3, ctrl_state = '111')
gc.append(mcz, [0,3,6,7])
gc.draw("mpl")
```
