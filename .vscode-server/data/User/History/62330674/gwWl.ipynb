```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Operator

qc = QuantumCircuit(2)
qc.x(0)
qc.s(1)
qc.cx(1,0)
qc.h(0)
qc.h(1)

#display(qc.draw("mpl"))
display(Operator.from_circuit(qc).draw("latex"))
```

```python
u = Operator([-2, 1])
v = Operator([1, 0, 1])
w = Operator([4, 1])
x = Operator([1/2, 1/2, 1/2])
psi = (u^v) - ((2/3)*w^x)
display(psi.draw("latex"))
```
