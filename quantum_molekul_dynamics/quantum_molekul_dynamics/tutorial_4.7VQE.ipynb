```python
import numpy as np
from qiskit.circuit.library import EfficientSU2
from qiskit.quantum_info import SparsePauliOp
from qiskit.primitives import StatevectorEstimator   # ⬅️ ganti ini
from scipy.optimize import minimize

hamiltonian = SparsePauliOp.from_list([
    ("ZI", 0.5),
    ("IZ", 0.5),
    ("XX", 0.2)
])

# ⬇️ estimator baru (tanpa shots)
estimator = StatevectorEstimator()

# Estimate energy
def energy(params, ansatz, hamiltonian, estimator):
    job = estimator.run([(ansatz, hamiltonian, params)])
    result = job.result()
    current_energy = result[0].data.evs  # ⬅️ akses hasil baru
    print("Energy: " + str(current_energy))
    return current_energy

# Ansatz
ansatz = EfficientSU2(hamiltonian.num_qubits)

initial_params = np.random.random(ansatz.num_parameters)

# Minimization
res = minimize(
    energy,
    initial_params,
    args=(ansatz, hamiltonian, estimator),
    method="COBYLA"
)

# Final result
final_params = res.x
print("Final Energy Estimate: " + str(energy(final_params, ansatz, hamiltonian, estimator)))
```
