```python
import math
from math import gcd
import random
```

```python
def shor_algo(N):
    #implementasi algoritma shor untuk faktorisasi bilangan N

    #cek apakah N adalah bilangan genap
    if N % 2 == 0:
        return 2, N//2

    #pilih bilangan acak a antara 2 dan N-1
    a = random.randint(2, N-1)

    #hitung gcd(a, N)
    g = gcd(a, N)
    if g!= 1:
        return g, N//g

    #menemukan order r dari a^x modulo N
    #untuk percobaan, kita akan menggunakan pendekatan klasik 
    r = 1
    result = a% N
    while result != 1:
        result = (result * a) % N
        r += 1
        if r > 10000: #batas untuk menghindari loop tak berujung
            break

    #jika r adalah ganjil, coba lagi
    if r % 2 == 1:
        return None

    #menghitung faktor
    x = pow(a, r//2, N)
    factor1 = gcd((x - 1)%N, N)
    factor2 = gcd((x + 1)%N, N)

    if factor1 == 1 or factor1 == N:
        return factor1, N//factor1
    if factor2 == 1 or factor2 == N:
        return factor2, N//factor2

    return None

#contoh penggunaan
N = 389648965298375023568

factors = shor_algo(N)
if factors:
    print(f'Factors of {N}: {factors[0]} dan {factors[1]}')
else:
    print(f'Failed to find factors')
```
